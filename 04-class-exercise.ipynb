```python
import numpy as np

#1. Open the file './data/befkbhalderstatkode.csv'
#2. Turn the csv file into a numpy ndarray with `np.genfromtxt(filename, delimiter=',', dtype=np.uint, skip_header=1)`
import pylab as pl

my_data = np.genfromtxt('./data/befkbhalderstatkode.csv', delimiter=',', dtype=np.uint, skip_header=1)

#3. Using this data: Find out how many people lived in each of the 11 areas in 2015
neighb = {1: 'Indre By', 2: 'Østerbro', 3: 'Nørrebro', 4: 'Vesterbro/Kgs. Enghave',
          5: 'Valby', 6: 'Vanløse', 7: 'Brønshøj-Husum', 8: 'Bispebjerg', 9: 'Amager Øst',
          10: 'Amager Vest', 99: 'Udenfor'}


for key, value in neighb.items():
    mask = (my_data[:,1] == key)
    mask2 = my_data[mask][:,4]
    total = np.sum(mask2)
    print(f'{value}: {total}')


#4. Make a bar plot to show the size of each city area from the smallest to the largest in 2015
import matplotlib.pyplot as plt
plt.bar(neighb.values(), list(set(my_data[:,1])))

#5. Create a boolean mask to find out how many people above 65 years lived in Copenhagen in 2015




```
