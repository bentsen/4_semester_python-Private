```python
import requests
import bs4 as bs
import re

#1. Get data from: "https://en.wikisource.org/wiki/Portal:State_of_the_Union_Speeches_by_United_States_Presidents"
url = "https://en.wikisource.org/wiki/Portal:State_of_the_Union_Speeches_by_United_States_Presidents"
req = requests.get(url)

#2. Using BeatifullSoup get all the speeches from 1900-2022
soup  = bs.BeautifulSoup(req.content, 'lxml')
speeches = soup.find_all('li')[144:269]
#print(speeches)



#3. Load all speech urls into a dictionary with year as key
speech_url = []
year_list = []

for li in speeches:
    speech_url.append("https://en.wikisource.org" + li.a.get('href'))
    this_date = li.find_all('a')[1].next_sibling
    year = re.search(r"\b(19|20)\d{2}\b", this_date)
    year_list.append(year.group(0))

speech_year_dict = {year_list[i]: speech_url[i] for i in range(len(year_list))}

#print(speech_year_dict)

#4. Loop through dictionary and save content of each speech in [year].txt files
for key, value in speech_year_dict.items():
    with open("year_files/" + key + ".txt", "w") as convert_file:
        print(f'Writing content from {value} to year_files/{key}')
        req = requests.get(value)
        soup = bs.BeautifulSoup(req.text, 'html')
        text = soup.select('div > p')[:-1]
        for lort in text:
            convert_file.write(lort.text)




```
